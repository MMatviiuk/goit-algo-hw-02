**Завдання 1.** Реалізація черги для системи обробки заявок у сервісному центрі

```python
import queue
import time
import random

# Створення черги заявок
черга_заявок = queue.Queue()
лічильник_заявок = 0

def створити_заявку():
    """Генерує нову заявку та додає її до черги."""
    global лічильник_заявок
    лічильник_заявок += 1
    заявка = f"Заявка_{лічильник_заявок}"  # Унікальний номер заявки
    черга_заявок.put(заявка)  # Додає заявку до черги
    print(f"Створено: {заявка}")

def обробити_заявку():
    """Обробляє заявку з черги, якщо вона є."""
    if not черга_заявок.empty():
        заявка = черга_заявок.get()  # Витягує заявку з черги
        print(f"Обробка: {заявка}")
        # Імітація часу обробки
        time.sleep(random.uniform(1, 3))
    else:
        print("Немає заявок для обробки")

# Головний цикл програми
if __name__ == "__main__":
    try:
        while True:
            створити_заявку()  # Генерує нову заявку
            обробити_заявку()  # Обробляє заявку
            time.sleep(1)  # Затримка між обробкою заявок
    except KeyboardInterrupt:
        print("Програму завершено.")
```

**Завдання 2.** Функція, яка приймає рядок як вхідний параметр, додає всі його символи до двосторонньої черги (deque з модуля collections в Python), а потім порівнює символи з обох кінців черги, щоб визначити, чи є рядок паліндромом.

```python
from collections import deque

def is_palindrome(input_string):
    # Видаляємо пробіли та перетворюємо рядок до нижнього регістру
    cleaned_string = ''.join(char.lower() for char in input_string if char.isalnum())

    # Створюємо двосторонню чергу
    char_deque = deque(cleaned_string)

    # Порівнюємо символи з обох кінців черги
    while len(char_deque) > 1:
        if char_deque.popleft() != char_deque.pop():
            return False

    return True

# Приклади використання
print(is_palindrome("racecar"))  # True
print(is_palindrome("hello"))  # False
```

**Завдання 3.** Програма, яка читає рядок з послідовністю символів-розділювачів, наприклад, ( ) { [ ] ( ) ( ) { } } }, і надає відповідне повідомлення, коли розділювачі симетричні, несиметричні, наприклад ( ( ( ) , або коли розділювачі різних видів стоять у парі, як-от ( }.

```python
def check_balanced(expression):
    """Перевіряє симметричність дужок використовуючі структуру Stack"""
    stack = []
    opening = "([{"
    closing = ")]}"
    mapping = {")": "(", "}": "{", "]": "["}

    for char in expression:
        if char in opening:
            stack.append(char)
        elif char in closing:
            if not stack or stack[-1] != mapping[char]:
                return "Несиметричні"
            stack.pop()

    return "Симетричні" if not stack else "Несиметричні"

expression = "() {} [] ( ) ( ) { }"
print(check_balanced(expression))  # Симетричні

expression = "( ( ( )"
print(check_balanced(expression))  # Несиметричні

expression = "( }"
print(check_balanced(expression))  # Несиметричні

expression = ")("
print(check_balanced(expression))  # Несиметричні (помилка: дужка, що закриває, йде першою)
```
